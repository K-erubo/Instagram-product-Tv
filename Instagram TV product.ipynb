# Import relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# Load the data

```python
data = pd.read_csv('2.01.Admittance.csv')
data
```

# Create data frame

```python
df = pd.DataFrame(data)
```

# Get average metrics

```python
average_views = data['views'].mean()
average_watch_time_minutes = data['watch_time_minutes'].mean()
average_likes = data['likes'].mean()
average_comments = data['comments'].mean()
average_shares = data['shares'].mean()
```

# Plot engagement metrics

```python
metrics = ['Average Views', 'Average Watch Time (minutes)', 'Average Likes', 'Average Comments', 'Average Shares']
values = [average_views, average_watch_time_minutes, average_likes, average_comments, average_shares]
```

# Let's plot data

```python
plt.figure(figsize=(10, 6))
plt.barh(metrics, values, color='purple')
plt.xlabel('Average Engagement')
plt.title('IGTV User Engagement Metrics')
plt.gca().invert_yaxis()
plt.show()
```
